```python
import pandas as pd 
```

```python
dataset=pd.read_csv("50_Startups.csv") 
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split         
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor    #here select model and train the model with training set
regressor=DecisionTreeRegressor(criterion='mae', splitter='best',max_features='auto')
regressor=regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_predict=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score         
r_score=r2_score(y_test,y_predict)
```

```python
r_score
```

```python
import pickle   
filename="Finalised model_decisionTree.sav"
pickle.dump(regressor,open(filename,'wb')) 
```
